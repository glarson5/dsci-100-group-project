### Data Wrangling and Cleaning

```r
# Count the number of sessions for each player
session_counts <- sessions_data |>
  filter(!is.na(end_time)) |>  # Remove incomplete sessions
  group_by(hashedEmail) |>
  summarise(num_sessions = n())

# Join session counts with player data and clean for analysis
analysis_data <- players_data |>
  left_join(session_counts, by = "hashedEmail") |>
  mutate(
    num_sessions = replace_na(num_sessions, 0),  # Players with no sessions get 0
    Age = as.numeric(Age)  # Convert Age to numeric
  ) |>
  filter(!is.na(Age))  # Remove players with missing age data

# Display the cleaned dataset structure
cat("Cleaned dataset contains", nrow(analysis_data), "players with complete age data\n")
cat("Players with at least one session:", sum(analysis_data$num_sessions > 0), "\n")
```
